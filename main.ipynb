```python
from ultralytics import YOLO
from ultralytics import SAM

yolo_model = YOLO("/content/runs/detect/train/weights/best.pt")

results = yolo_model("/content/0118_JPEG_jpg.rf.1f0122577672c539d444e12c01eff54d.jpg")

sam_model = SAM("sam2_b.pt")

for result in results:
  class_ids = result.boxes.cls.int().tolist()
  if len(class_ids):
    boxes = result.boxes.xyxy
    masks = sam_model(result.orig_img, bboxes= boxes, verbose=False, save=True, device=0)
```
